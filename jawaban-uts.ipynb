### 1. Jelaskan Apa itu Machine Learning

**Machine Learning** (Pembelajaran Mesin) adalah cabang dari kecerdasan buatan (Artificial Intelligence, AI) yang berfokus pada pengembangan algoritma dan teknik yang memungkinkan komputer untuk belajar dari dan membuat prediksi atau keputusan berdasarkan data. Secara sederhana, machine learning memungkinkan sistem komputer untuk meningkatkan kinerja tugas tertentu melalui pengalaman tanpa perlu diprogram secara eksplisit untuk tugas tersebut.

### 2. Contoh Penerapan Implementasi dalam Kehidupan Sehari-hari

### Sistem Rekomendasi (Recommendation Systems)
Layanan seperti Netflix, Spotify, dan Amazon menggunakan algoritma ML untuk merekomendasikan film, musik, atau produk kepada pengguna berdasarkan riwayat penggunaan mereka.

- **Alasan dan Manfaatnya:**
  - personalisasi: Membuat pengalaman pengguna lebih personal dan relevan.
  - Efisiensi: Menghemat waktu pengguna dalam mencari konten yang mereka sukai.
  - Peningkatan Penjualan: Membantu platform dalam meningkatkan penjualan atau engagement     dengan merekomendasikan produk atau konten yang lebih mungkin untuk diminati oleh pengguna.


### Asisten Virtual (Virtual Assistants)
Siri, Google Assistant, dan Alexa menggunakan ML untuk memahami dan merespons perintah suara pengguna.

- **Alasan dan Manfaaatnya:**
  - Kemudahan Akses: Memudahkan pengguna untuk mengakses informasi dan layanan hanya dengan perintah suara.
  - Peningkatan Produktivitas: Membantu dalam mengatur jadwal, mengirim pesan, dan melakukan berbagai tugas sehari-hari dengan lebih efisien.


### 3. Macam-macam Taxonomi dalam Penerapan Machine Learning

### Supervised Learning (Pembelajaran Terawasi)
- **Definisi :** Model dilatih menggunakan dataset berlabel, yang berarti setiap input disertai dengan output yang diinginkan.
- **Contoh :** Klasifikasi email sebagai "spam" atau "bukan spam", prediksi harga rumah berdasarkan fitur-fitur tertentu.
- **Manfaat :** Memungkinkan prediksi yang akurat ketika ada data yang cukup dan berlabel.

### (Unsupervised Learning) Pembelajaran Tak Terawasi 
- **Definisi :** Model dilatih menggunakan dataset yang tidak berlabel, dan model mencoba menemukan struktur atau pola dalam data.
- **Contoh :** Klasterisasi pelanggan berdasarkan perilaku belanja, pengurangan dimensi data.
- **Manfaat :** Berguna untuk menemukan pola tersembunyi atau segmentasi dalam data tanpa intervensi manusia.

### (Semi-Supervised Learning) Pembelajaran Semi-Terawasi 
- **Definisi :** Kombinasi antara pembelajaran terawasi dan tak terawasi, di mana sebagian data diberi label dan sebagian tidak.
- **Contoh :** Klasifikasi teks di mana hanya sebagian kecil dokumen yang diberi label.
- **Manfaat :** Menggunakan sejumlah kecil data berlabel untuk meningkatkan akurasi model dengan data tak berlabel yang melimpah.

### (Reinforcement Learning) Pembelajaran Penguatan
- **Definisi :** Model belajar melalui trial and error di lingkungan yang dinamis dengan mendapatkan umpan balik berupa reward atau punishment.
- **Contoh :** Robotik yang belajar berjalan, algoritma bermain game seperti catur atau Go.
- **Manfaat :** Efektif untuk masalah di mana keputusan harus dibuat dalam urutan yang terstruktur dan melibatkan interaksi dengan lingkungan.

```python
import pandas as pd

df = pd.read_csv('dataset-soalsk-pertama.csv')

df.columns = df.columns.str.strip()

df['Biaya'] = df['Biaya'].apply(lambda x: int(x.split('x')[0]) * int(x.split('x')[1]))

rata_rata_datang = df['Datang'].mean()
print("a) Rata-rata mahasiswa datang pada minggu ini :", rata_rata_datang)

biaya_tertinggi = df.loc[df['Biaya'].idxmax()]
print("b) Biaya tertinggi terjadi pada hari :", biaya_tertinggi['Hari'])

hari_biaya_lebih_110000 = df.loc[df['Biaya'] > 110000, 'Hari'].tolist()
hari_biaya_lebih_110000_str = ', '.join(hari_biaya_lebih_110000)
print("c) Hari apa biaya lebih dari 110000 :", hari_biaya_lebih_110000_str)

mahasiswa_paling_banyak_datang = df['Mahasiswa'].mode().tolist()
mahasiswa_paling_banyak_datang_str = ', '.join(mahasiswa_paling_banyak_datang)
print("d) Mahasiswa yang paling banyak datang ke kampus :", mahasiswa_paling_banyak_datang_str)

mahasiswa_hari_minggu = df.loc[df['Hari'] == 'Minggu', 'Mahasiswa'].tolist()
mahasiswa_hari_minggu_str = ', '.join(mahasiswa_hari_minggu)
print("e) Mahasiswa yang datang pada hari Minggu :", mahasiswa_hari_minggu_str)

biaya_tertinggi = df['Biaya'].max()
biaya_terendah = df['Biaya'].min()
print("f) Biaya tertinggi:", biaya_tertinggi, "& Biaya terendah :", biaya_terendah)

frekuensi_datang_tertinggi = df['Mahasiswa'].value_counts().max()
frekuensi_datang_terendah = df['Mahasiswa'].value_counts().min()
print("g) Frekuensi datang tertinggi :", frekuensi_datang_tertinggi, "& Frekuensi datang terendah :", frekuensi_datang_terendah)
```
